```python
%matplotlib inline

import numpy as np
import keras as keras
```

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
x_train2 = np.reshape(x_train, (60000, 784))
# Ces deux lignes là sont super importante ! Si on les met pas le modèle fait de la merde
x_train2 = x_train2.astype('float32')
x_train2 /= 255
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='sgd',loss='categorical_crossentropy', metrics=['accuracy'])

```

```python
from keras.utils.np_utils import to_categorical
y_train2 = to_categorical(y_train, 10)
```

```python
model.fit(x_train2, y_train2, validation_split=0.2, nb_epoch=4)
```
